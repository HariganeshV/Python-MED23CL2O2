STEPS:

1.Create a empty result list

2.Keep a total variable starting at 0

3.Loop through each number

4.Add to total , append total to result

**Running sum**

```python
#Running sum   Sample ip=10,20,5,1  Expected op=10,30,35,36
def runningSum(nums):
  result = [] #Empty list
  total = 0 #Initialist total to zero
  for n in nums: #start the loop for running sum
    total += n #add the value to total
    result.append(total) #append the sum to result list
  return(result)
print(runningSum([10,20,5,1]))
```

**Count Even number in a list**

```python
#Count Even number in a list
#Count Even number
#Sample input (3,7,2,9,5)
#Expected output : 1

def CountEven(nums):
  count = 0
  for n in nums:
    if n%2==0:
      count +=1
  return count
print(CountEven([3,7,2,9,5]))
```

**Reverse a list**

**Find Maximum number in the list**

```python
# #Find Maximum number in the list
# input[3,7,2,9,5]
# Expected output:9

def FindMax(nums):
  max_num = nums[0]#Assign first of list as maximum
  for n in nums:
    if n>max_num:
      max_num = n
  return max_num
print(FindMax([3,7,2,9,5]))
```

**Find Minimum number in the list**

```python
# #Find Minimum number in the list
# input[3,7,2,9,5]
# Expected output:2

def FindMin(nums):
  min_num = nums[0]#Assign first of list as minimum
  for n in nums:
    if n<min_num:
      min_num = n
  return min_num
print(FindMin([3,7,2,9,5]))
```

**Merge 2 sorted list**

```python
#Merge 2 sorted list
#Input: [1,3,5],[2,4,6]
#Output: [1,2,3,4,5,6]

def mergeList(list1,list2):
  return sorted(list1 + list2)
print(mergeList([1,3,5],[2,4,6]))
```

**Square of a sorted list**

```python
#Square of a sorted list

def SortedSquare(nums):
  return sorted([x**2 for x in nums])
# List comprehension: Defines a list based on iteration that is
#[x**2 for x in nums]
#is defined to find square of elements in the sorted list
print(SortedSquare([-4,-1,0,3,-50]))
```

```python
def MaxDeposit(accounts):
  return max(sum(cust) for cust in accounts)
print(MaxDeposit([[100,52],[3,60]]))
```

**Find 2 indices where numbers add up to the target**

```python
# Find 2 indices where numbers add up to the target
#Input: [2,7,11,15] target:26
#Expected Output: [2,3]

def twoSum(nums,target):
  #define a dictionary , a key value pair
  seen ={}
  for i,n in enumerate(nums):
    diff = target - n
    if diff in seen:
       return [seen[diff],i]
    seen[n] = i
    print(seen)
print(twoSum([2,7,11,15],26))
```

```python
class CoffeeMachine:
  def __init__(self):
    self.water = 100
  def make_coffee(self):
      if self.water >= 10:
        self.water -= 10
        print("Coffee is ready")
      else:
        print("not enough water to make coffee")
machine=CoffeeMachine()
machine.make_coffee()
```

#Encapsulation

```python
class BankAccount:
  def _init_(self, amount):
    self.__balance = 0
  def set_balance(self, amount):
    self.__balance = amount
  def get_balance(self):
    return self.__balance
account = BankAccount()
account.set_balance(1000)
print("Account balance:", account.get_balance())
```

#Inheritance

Multi level

```python
class Animal:
  def move(self):
    print("Animal moves")

class Bird(Animal):
  def fly(self):
    print("Bird flies")

class Parrot(Bird):
  def talk(self):
    print("Parrot talks")

parrot = Parrot()
parrot.move()
parrot.fly()
parrot.talk()
```

Single

```python
class Father:
  def speak(self):
    print("Father speaks")

class Son(Father):
  def recieve(self):
    print("Son recieves")

son = Son()
son.speak()
son.recieve()
```

Polymorphism

```python
class cat:
  def sound(self):
    print("meow")
class dog:
  def sound(self):
    print("bark")
for animal in (cat(), dog()):
  animal.sound()
```

#10 Nov 2025

User Definition Exception

```python
class NegativeNumberError(Exception):
    """Raised when a negative number is entered"""
    pass

def check_number(num):
    if num < 0:
        raise NegativeNumberError("Negative number not allowed!")
    else:
        print(f"Square of {num} is {num**2}")

try:
    n = int(input("Enter a positive number: "))
    check_number(n)
except NegativeNumberError as e:
    print("Custom exception:", e)
except ValueError:
    print("Please enter a valid integer")
```

```python
class AgeToSmallError(Exception):
  """Raised when age is less than 18."""
  pass

try:
  age = int(input("Enter your age:"))
  if age < 18:
    raise AgeToSmallError("You must be at least 18 years old.")
  print("Access granted.")
except AgeToSmallError as e:
  print("Custom Exception:",e)
except ValueError:
  print("Invalid Input.Please enter a number.")
finally:
  print("Thank you for using our system")
```
